## Predicting party affiliation from voting records
### Abstract
How members of the Democratic and Republican parties vote is a strong indicator of a particular political climate. In this [dataset, provided by the UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/congressional+voting+records), votes on 16 key issues by the 1984 98th House of Representatives were identified as yea, nay, or unknown. Given that some issues are highly partisan (e.g. 98% of Republicans and 7% of Democrats) voted for "physician-fee-freeze", I decided to compare my machine learning models to a benchmark model where all yea votes for this issue were classified as Republicans, and all nay votes as Democrats. This led to a balanced accuracy (due to there being more Democrats than Republicans; accuracy hereafter) of 95.9% . In exploratory analysis, I ran a two component PCA, and indeed, the first PC separated observations along party lines. Next, I ran three machine learning models, a logistic regression, a decision tree, and a random forest. The decision tree ended up being logically identical to the benchmark model. The logistic regression and random forest models outperformed the benchmark model, with 96.6% and 98.5% accuracy in the test set. Although all models heavily relied on the most partisan issue, "physician-fee-freeze," the more sophisticated machine learning models that included extra features were more successful at categorizing anomolous Democrats and Republicans who do not strictly vote along party lines. In the future, I am interested in utilizing a similar analytic paradigm to look at a [larger and more contemporary dataset, provided by the federal government](https://www.govtrack.us/congress/votes). This dataset also includes the names of the politicians, and it would be interesting to formally identify the most "classifiable" and "unclassifiable" politicians, and the issues that unite and divide us today.
### Introduction
This dataset has 435 observations (for each member of the House of Representatives), 16 features (votes on 16 different issues), and the target political affiliation. It is very important to note that the column names are ambiguous. At first glance, it looks like "aid-to-nicaraguan-contras" (the Contras being a right-wing rebel group in Nicaragua) was a heavily Democrat-leaning issue (82% Democrats for vs. 18% Republicans for). However, the actual bill had to do with ["Authorizations for Intelligence Agencies Ban Aid to Contras."](https://library.cqpress.com/cqalmanac/toc.php?mode=cqalmanac-toc&level=3&values=1984+-+98th+Congress%2C+2nd+Session%7EChapter+3+-+Foreign+Policy) In addition, it looks like "el-salvador-aid" was a heavily Republican-leaning issue (23% Democrats for vs. 94% Republicans for) of but the crux of the bill was not just about providing aid, but providing aid while ["tying the funds to progress there on human rights."](https://www.washingtonpost.com/archive/politics/1984/05/11/reagan-wins-narrowly-on-aid-to-el-salvador/6ed8ac74-99b1-4fcc-8d9d-c57ba1295891/) Furthermore, "physician-fee-freeze" sounds like a measure that would lead to patients spending less money on physician fees for healthcare (and thereby doctors making less money). Historically, it does not make sense that this would be a heavily Republican-leaning issue (7% Democrats for vs. 98% Republicans). I could not find a definative answer to this. The actual measure could be [an attempt to reduce government Medicare costs.](https://www.nytimes.com/1984/04/13/us/house-votes-cuts-in-benefit-plans-to-reduce-deficit.html) It could also be regarding [whether the freeze is voluntary versus mandatory, or whether it applies to inpatients vs. outpatients.](https://www.govinfo.gov/content/pkg/GPO-CRECB-1984-pt7/pdf/GPO-CRECB-1984-pt7-2-2.pdf)

I was extremely skeptical and confused about this dataset at first, but after checking the voting records against primary sources, I feel confident that it is accurate. I will make sure that my conclsions are appropriate with regards to the nuances of historical context.
### Research questions
* What were the issues that most separated Democrats and Republicans in this Congress?

### Methods 1.1: Load data

```python
# Read csv
import pandas as pd
csv = pd.read_csv('Data/Congress.csv')
print(csv.shape)
```

```python
# Create DataFrame
data = pd.DataFrame(data = csv)
data.head()
```

### Methods 1.2: Feature engineering
For simplicity, I will code unknown/? as between yes and no, i.e., 0.5. Later I might try to treat yes, no, and unknown as
categorical variables and quantify them with one-hot encoding.

```python
# Quantify votes
data = data.replace(to_replace = 'y', value = 1)
data = data.replace(to_replace = 'n', value = 0)
data = data.replace(to_replace = '?', value = 0.5)
# Quantify democrat as 1 and republican as 0
data = data.replace(to_replace = 'democrat', value = 1)
data = data.replace(to_replace = 'republican', value = 0)
# Convert to float
data = data.astype('float')
```

### Methods 2.1: Exploratory data analysis
* What was the composition of this Congress?
* What were the yea-to-nay vote counts on each issue?

```python
print('There were', data['party'].sum(), 'Democrats and', len(data) - data['party'].sum(), 
      'Republicans in this Congress')
print(round(100*data['party'].sum()/len(data), 1), '% of this Congress were Democrats')
for column in data.iloc[:,1:]:
    print('The vote on', column, 'was', data[column].sum(), 'to', len(data) - data[column].sum())
```

### Methods 2.2: Exploratory data analysis
* Which party had a higher standard deviation in their voting record?

```python
data.groupby('party').std().round(2)
```

```python
print('Democrat average standard deviation is', round(data.groupby('party').std().iloc[0].mean(), 2))
print('Republican average standard deviation is', round(data.groupby('party').std().iloc[1].mean(), 2))
```

In this particular Congress, Republicans have more variability in their voting record than Democrats

### Methods 2.3: Exploratory data analysis
* What were the most partisan issues in this Congress?

```python
data.groupby('party').mean().round(2)
```

```python
# Examine most partisan issues (positive = Democrat-leaning)
partisan = data.groupby('party').mean().round(2).iloc[1] - data.groupby('party').mean().round(2).iloc[0]
# Order by 
partisan.sort_values()
```

Again, as I emphasized in the introduction, the names of these features are very ambiguous and must be interpreted with care.

### Methods 2.4: Principal component analysis

```python
# Principal component dimensionality reduction
from sklearn.decomposition import PCA
import seaborn as sns
pca = PCA(n_components = 2)
pca.fit(data.iloc[:,1:].T)
pca_df = pd.DataFrame(pca.components_)
pca_df = pca_df.T
pca_df['party'] = data['party']
sns.scatterplot(pca_df[0], pca_df[1], hue = pca_df['party'])
```

Democrats and Republicans seem to be segregated by PC1 (the component that explains the most variance). What issues are correlated with PC1 and what does PC2 represent?

```python
data['pc1'] = pca_df[0]
data['pc2'] = pca_df[1]
data.corr()['pc1'].sort_values()
```

Interestingly, there are features that correlate more strongly with PC1 than party. It looks like the features have more to do with having a right-wing stance on foreign policy (eg. El Salvador, Nicaragua, missiles)

```python
data.corr()['pc2'].sort_values()
```

For PC2, looks like the features have more to do with having a left-wing stance on domestic policy (water, fuels, immigration)

In terms of the scatterplot, PC1 is heavily party-based, with positive values indicating Republican right-wing foreign policy views. PC2 is less strictly party-based, but does slightly lean left (correlation with party = 0.22), and seems to indicate left-wing domestic policy views. 

### Methods 3.1: Benchmark model
Given that 98% of Republicans voted for "physician-fee-freeze," we can imagine a simplistic benchmark model where every representative who voted for it is classified as Republican and every representative who did not vote for it is classified as Democrat. If a machine learning model cannot perform better than this, it should not be considered to be very impressive.

```python
data['dummy-democrat'] = ~(data['physician-fee-freeze'] == 1)
```

```python
# Plot confusion matrix
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, balanced_accuracy_score, confusion_matrix
cm = confusion_matrix(data['party'], data['dummy-democrat'])
plt.figure(figsize = (10,10))
sns.heatmap(cm, annot = True, fmt = ".0f", square = True, annot_kws = {"fontsize": 50});
plt.ylabel('Actual label', fontsize = 20);
plt.xlabel('Predicted label', fontsize = 20);
score = accuracy_score(data['party'], data['dummy-democrat']).round(2)
title = 'Accuracy Score: {0}'.format(score)
plt.title(title, size = 20);
plt.show()
# Other methods of scoring
print('Accuracy = ', accuracy_score(data['party'], data['dummy-democrat']).round(3))
print('Precision =', precision_score(data['party'], data['dummy-democrat']).round(3))
print('Recall =', recall_score(data['party'], data['dummy-democrat']).round(3))
print('F1 Score =', f1_score(data['party'], data['dummy-democrat']).round(3))
print('Balanced accuracy =', balanced_accuracy_score(data['party'], data['dummy-democrat']).round(3))
# Balanced accuracy of dummy model is 95.9%
```

### Methods 4.1: Prepare data for machine learning

```python
# Separate features from target
X = data.drop(['party', 'pc1', 'pc2', 'dummy-democrat'], axis = 1)
y = data['party']
```

```python
# Examine multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame(columns = ['Features', 'VIF'])
vif['Features'] = X.columns
for i in range(X.shape[1]):
    vif.loc[i, 'VIF'] = variance_inflation_factor(X.values, i).round(1)
# Remove constant row
vif = vif[1:]
# Sort by VIF
vif.sort_values('VIF', ascending = False)
```

Lots of multicollinearity, as expected by correlated voting by party

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 20201027, stratify = y)
print(X_train.shape)
print(X_test.shape)
```

### Methods 4.2: Logistic regression with cross-validation and grid search

```python
# Logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
import numpy as np
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
grid = {'C': [0.01, 0.05, 0.1, 0.5, 1, 5, 10], 'penalty' :['elasticnet'], 'solver': ['saga'], 
        'l1_ratio': [0, 0.2, 0.4, 0.6, 0.8, 1], 'class_weight': ['balanced', None]}
gridsearch = GridSearchCV(estimator = logreg, param_grid = grid, n_jobs = -1,                      
                          cv = 10, return_train_score = True, scoring = 'balanced_accuracy')
gridsearch.fit(X_train, y_train)
logreg_models = pd.DataFrame(gridsearch.cv_results_).sort_values('mean_test_score', ascending = False)
print('LR mean test score:', round(logreg_models['mean_test_score'].iloc[0], 3))
print('LR best parameters', logreg_models['params'].iloc[0]) # Best parameters, 100% lasso regularization
```

```python
# Train with best paraeters
logreg = LogisticRegression(C = 1,
 class_weight = 'balanced',
 l1_ratio = 1,
 penalty = 'elasticnet',
 solver = 'saga')
logreg.fit(X_train, y_train)
```

```python
# Sort features by beta value
feature_coefs = pd.DataFrame(columns = ['Feature', 'Beta'])
feature_coefs['Feature'] = X.columns
coefs = logreg.coef_.round(2)
feature_coefs['Beta'] = coefs[0]
feature_coefs.sort_values('Beta')
```

As expected, "physician-fee-freeze," the most partisan issue, also has the largest beta value.

```python
# Plot confusion matrix
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, balanced_accuracy_score, confusion_matrix
y_pred_logreg = logreg.predict(X_train)
cm = confusion_matrix(y_train, y_pred_logreg)
plt.figure(figsize = (10,10))
sns.heatmap(cm, annot = True, fmt = ".0f", square = True, annot_kws={"fontsize": 50});
plt.ylabel('Actual label', fontsize = 20);
plt.xlabel('Predicted label', fontsize = 20);
score = accuracy_score(y_train, y_pred_logreg).round(2)
title = 'Accuracy Score: {0}'.format(score)
plt.title(title, size = 20);
plt.show()
# Other methods of scoring
print('===Logistic regression===')
print('Accuracy = ', accuracy_score(y_train, y_pred_logreg).round(3))
print('Precision =', precision_score(y_train, y_pred_logreg).round(3))
print('Recall =', recall_score(y_train, y_pred_logreg).round(3))
print('F1 Score =', f1_score(y_train, y_pred_logreg).round(3))
print('Balanced accuracy =', balanced_accuracy_score(y_train, y_pred_logreg).round(3))
```

Logistic regression has a 96.5% balanced accuracy (cross-validation score) vs. the 95.9% of the dummy model, an improvement of 0.6%

### Methods 4.3: Decision tree

```python
# Decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state = 20201027)
grid = {'max_depth': [None, 1, 5, 10], 'max_features': ['sqrt', None], 'class_weight': ['balanced', None]}
gridsearch = GridSearchCV(estimator = dtc, param_grid = grid, n_jobs = -1,                      
                          cv = 10, return_train_score = True, scoring = 'balanced_accuracy')
gridsearch.fit(X_train, y_train)
dtc_models = pd.DataFrame(gridsearch.cv_results_).sort_values('mean_test_score', ascending = False)
print('LR mean test score:', round(logreg_models['mean_test_score'].iloc[0], 3))
print('LR mest parameters', logreg_models['params'].iloc[0])
print('DTC mean test score:', round(dtc_models['mean_test_score'].iloc[0], 3))
print('DTC best parameters', dtc_models['params'].iloc[0])
```

```python
# Train with best parameters
dtc = DecisionTreeClassifier(class_weight = None, max_depth = 1, max_features = None, 
                             random_state = 20201027)
dtc.fit(X_train, y_train)
```

```python
# Plot tree
from sklearn.tree import plot_tree
fig = plt.figure(figsize = (100, 20))
_ = plot_tree(dtc, 
                   feature_names = X.columns,  
                   class_names = ['Republican', 'Democrat'],
                   filled = True)
```

```python
# Plot confusion matrix
y_pred_dtc = dtc.predict(X_train)
cm = confusion_matrix(y_train, y_pred_dtc)
plt.figure(figsize = (10,10))
sns.heatmap(cm, annot = True, fmt = ".0f", square = True, annot_kws={"fontsize": 50});
plt.ylabel('Actual label', fontsize = 20);
plt.xlabel('Predicted label', fontsize = 20);
score = round(accuracy_score(y_train, y_pred_dtc),2)
title = 'Accuracy Score: {0}'.format(score)
plt.title(title, size = 20);
plt.show()
print('===Logistic regression===')
print('Accuracy = ', accuracy_score(y_train, y_pred_logreg).round(3))
print('Precision =', precision_score(y_train, y_pred_logreg).round(3))
print('Recall =', recall_score(y_train, y_pred_logreg).round(3))
print('F1 Score =', f1_score(y_train, y_pred_logreg).round(3))
print('Balanced accuracy =', balanced_accuracy_score(y_train, y_pred_logreg).round(3))
print('===Decision tree classifier===')
print('Accuracy = ', accuracy_score(y_train, y_pred_dtc).round(3))
print('Precision =', precision_score(y_train, y_pred_dtc).round(3))
print('Recall =', recall_score(y_train, y_pred_dtc).round(3))
print('F1 Score =', f1_score(y_train, y_pred_dtc).round(3))
print('Balanced accuracy =', balanced_accuracy_score(y_train, y_pred_dtc).round(3))
# Decision tree classifier has a 95.7% balanced accuracy vs. the 95.9% of the dummy model, which is worse
```

Not surprisingly, the decision tree classifier 100% replicates the benchmark model

### Methods 4.4: Random forest

```python
# Random forest classifier
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state = 20201027)
grid = {'max_depth': [None, 1, 5, 10], 'class_weight': ['balanced', None]}
gridsearch = GridSearchCV(estimator = rfc, param_grid = grid, n_jobs = -1,                      
                          cv = 10, return_train_score = True, scoring = 'balanced_accuracy')
gridsearch.fit(X_train, y_train)
rfc_models = pd.DataFrame(gridsearch.cv_results_).sort_values('mean_test_score', ascending = False)
print('LR mean test score:', round(logreg_models['mean_test_score'].iloc[0], 3))
print('LR mest parameters', logreg_models['params'].iloc[0])
print('DTC mean test score:', round(dtc_models['mean_test_score'].iloc[0], 3))
print('DTC best parameters', dtc_models['params'].iloc[0])
print('RFC mean test score:', round(rfc_models['mean_test_score'].iloc[0], 3))
print('RFC best parameters', rfc_models['params'].iloc[0])
```

```python
# Train with best parameters
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(class_weight = 'balanced', max_depth = 10, random_state = 20201014)
rfc.fit(X_train, y_train)
```

```python
# Plot feature importance
from sklearn.inspection import permutation_importance
result = permutation_importance(rfc, X_train, y_train, n_repeats = 10,
                                random_state = 20201014, n_jobs = -1)
sorted_idx = result.importances_mean.argsort()
fig, ax = plt.subplots()
ax.boxplot(result.importances[sorted_idx].T,
           vert = False, labels = X.columns[sorted_idx])
ax.set_title("Permutation Importances")
fig.tight_layout()
plt.show()
```

Also not suprisingly, the random forest classifier considers 'physician-fee-freeze' the most important feature. However, due to using only sqrt(features) for each tree, it also uses some other features to classify party affiliation.

```python
# Plot confusion matrix
y_pred_rfc = rfc.predict(X_train)
cm = confusion_matrix(y_train, y_pred_rfc)
plt.figure(figsize = (10,10))
sns.heatmap(cm, annot = True, fmt = ".0f", square = True, annot_kws={"fontsize": 50});
plt.ylabel('Actual label', fontsize = 20);
plt.xlabel('Predicted label', fontsize = 20);
score = round(accuracy_score(y_train, y_pred_rfc),2)
title = 'Accuracy Score: {0}'.format(score)
plt.title(title, size = 20);
plt.show()
# Other methods of scoring
print('===Logistic regression===')
print('Accuracy = ', accuracy_score(y_train, y_pred_logreg).round(2))
print('Precision =', precision_score(y_train, y_pred_logreg).round(2))
print('Recall =', recall_score(y_train, y_pred_logreg).round(2))
print('F1 Score =', f1_score(y_train, y_pred_logreg).round(2))
print('Balanced accuracy =', balanced_accuracy_score(y_train, y_pred_logreg).round(2))
print('===Decision tree classifier===')
print('Accuracy = ', accuracy_score(y_train, y_pred_dtc).round(2))
print('Precision =', precision_score(y_train, y_pred_dtc).round(2))
print('Recall =', recall_score(y_train, y_pred_dtc).round(2))
print('F1 Score =', f1_score(y_train, y_pred_dtc).round(2))
print('Balanced accuracy =', balanced_accuracy_score(y_train, y_pred_dtc).round(2))
print('===Random forest classifier===')
print('Accuracy = ', accuracy_score(y_train, y_pred_rfc).round(2))
print('Precision =', precision_score(y_train, y_pred_rfc).round(2))
print('Recall =', recall_score(y_train, y_pred_rfc).round(2))
print('F1 Score =', f1_score(y_train, y_pred_rfc).round(2))
print('Balanced accuracy =', balanced_accuracy_score(y_train, y_pred_rfc).round(2))
```

100% accuracy! but this might indicate overfitting on training data.
